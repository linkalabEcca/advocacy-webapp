```python
import pandas as pd
import tldextract
```

```python
df = pd.read_excel('data/df_testate_complete&wiki.xlsx')
df = df.fillna(0)
#edited_df = st.data_editor(df, num_rows="dynamic")
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[['Virality_Frequency_News_quot1']] = scaler.fit_transform(df[['counts_Frequency_Topic_quotidiani_1']])
df[['Virality_Cluster_News_quot1']] = scaler.fit_transform(df[['counts_cluster_categ_1']])
df[['Virality_Topic_News_quot1']] = scaler.fit_transform(df[['counts_Trends_quotidiani_1_topic']])

df[['Virality_Frequency_News_quot2']] = scaler.fit_transform(df[['counts_Frequency_Topic_quotidiani_2']])
df[['Virality_Cluster_News_quot2']] = scaler.fit_transform(df[['counts_cluster_categ_2']])
df[['Virality_Topic_News_quot2']] = scaler.fit_transform(df[['counts_Trends_quotidiani_2_topic']])

df[['Virality_Frequency_News']] = scaler.fit_transform(df[['counts_Frequency_Topic']])
df[['Virality_Cluster_News']] = scaler.fit_transform(df[['counts_cluster_categ']])
df[['Virality_Topic_News']] = scaler.fit_transform(df[['counts_Trends_topic']])

df['Virality_Score_1.gnews'] = ((df['Virality_Topic_News'])  + (df['Virality_Topic_News_quot1']) + (df['Virality_Topic_News_quot2']) +(df['Virality_Cluster_News'])  + (df['Virality_Cluster_News_quot1']) + (df['Virality_Cluster_News_quot2'])+(df['Virality_Frequency_News'])  + (df['Virality_Frequency_News_quot1']) + (df['Virality_Frequency_News_quot2'])) 
df[['Virality_Score_1.gnews']] = scaler.fit_transform(df[['Virality_Score_1.gnews']])

df = df.rename(columns={'Score_wiki_keywords_semantic_scaled':'Virality_Score_2.wiki','Score_trends_keywords_semantic_scaled':'Virality_Score_3.gtrends'})

df['Virality_Score'] = (df['Virality_Score_1.gnews'] + df['Virality_Score_2.wiki'] + df['Virality_Score_3.gtrends'])/3
df[['Virality_Score']] = scaler.fit_transform(df[['Virality_Score']])

df['Final_Score'] = ((0.3 * df['Virality_Score'] + (0.7 * df['Complete_Score_manager']))/2)
df[['Final_Score']] = scaler.fit_transform(df[['Final_Score']])

df = df.drop_duplicates(subset=['title','domain']).drop_duplicates(subset=['summary'])
```

```python
df = df.sort_values('Final_Score',ascending = False)[['data','domain','title','summary','message','SimilarityScore_keywords_semantic','Complete_Score_manager','Virality_Score','Virality_Score_1.gnews','Virality_Score_2.wiki','Virality_Score_3.gtrends','Final_Score','Permalink','source','languagecode']].rename(columns={'title_x':'title'}).reset_index(drop=True)

def get_domain(x):
    x= x.replace('[','').replace(']','')
    print(x)
    extracted_info = tldextract.extract(x)
    domain = extracted_info.domain
    print(domain)
    return domain
    
df.domain = df.domain.apply(get_domain)
df.to_excel('data/dataset_app.xlsx')
```
